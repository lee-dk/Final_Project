```python
import pandas as pd
```

```python
#csv파일 dataframe으로 변환하고 읽기
review_data = pd.read_csv('./yogiyo/yogiyo_review_data_update.csv', 
                          encoding='ansi')
#불필요한 데이터 drop하기
review_data.drop('Unnamed: 0', axis=1, inplace=True)

review_data.head(16)
```

```python
#아이템기반(Main Menu) 협업 필터링_ID별
userID_review = review_data.pivot_table('point', 
                                        index = 'user_id', 
                                        columns='order_item').fillna(0)
userID_review
```

```python
#아이템기반(Main Menu) 협업 필터링_유저별
review_userID = review_data.pivot_table('point', 
                                        index = 'order_item', 
                                        columns='user_id').fillna(0)
review_userID
```

```python
#사이드메뉴까지 속성으로 두기엔 너무 방대하다...각자 시킨 메뉴가 다 다르기때문에 메인메뉴만 해야할 듯...?
#그러기 위해서는.......order_item에서 첫번째나오는 / 이후 내용들은 삭제하고싶은데..어떻게 하면 좋을까...
#위에가 해결된다면,,index를 메인 메뉴(16가지)로 두고 columns을 user_id로 한 다음 data를 point로 주면 메뉴추천 시스템으로 만들 수 있을 듯

#그래서 세분화필요..!order_item을 세분화시켜서 메인따로 사이드 따로 만들것!
```

```python
#코사인 유사도(Cosine Similarity) : userID_review
from sklearn.metrics.pairwise import cosine_similarity

user_based_collabo = cosine_similarity(userID_review)

user_based_collabo
```

```python
#코사인 유사도(Cosine Similarity)_review_userID
from sklearn.metrics.pairwise import cosine_similarity

item_based_collabo = cosine_similarity(review_userID)

item_based_collabo
```

```python
user_based_collabo = pd.DataFrame(data = user_based_collabo, 
                                 index = userID_review.index, 
                                 columns = userID_review.index)
user_based_collabo
```

```python
item_based_collabo = pd.DataFrame(data = item_based_collabo, 
                                 index = review_userID.index, 
                                 columns = review_userID.index)
item_based_collabo
```

```python
#추천 시스템_아이템기반
def get_item_based_collabo(order_item):
    return item_based_collabo[order_item].sort_values(ascending=False)[:6]
```

```python
get_item_based_collabo('홈메이드 감자크림스프')
```

```python
#추천 시스템_사용자기반
def get_user_based_collabo(user_id):
    return user_based_collabo[user_id].sort_values(ascending=False)[:6]
```
